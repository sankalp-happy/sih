```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('cleaned_file.csv')
```

Now defining the features variables (using the pre & post moonsoon data from 2015 to 2021 )

```python
features = df[['Pre-monsoon_2015 (meters below ground level)', 'Post-monsoon_2015 (meters below ground level)', 'Pre-monsoon_2016 (meters below ground level)', 'Post-monsoon_2016 (meters below ground level)', 'Pre-monsoon_2017 (meters below ground level)', 'Post-monsoon_2017 (meters below ground level)', 'Pre-monsoon_2018 (meters below ground level)', 'Post-monsoon_2018 (meters below ground level)', 'Pre-monsoon_2019 (meters below ground level)', 'Post-monsoon_2019 (meters below ground level)', 'Pre-monsoon_2020 (meters below ground level)', 'Post-monsoon_2020 (meters below ground level)', 'Pre-monsoon_2021 (meters below ground level)', 'Post-monsoon_2021 (meters below ground level)']]
```

Defining the target variables. (predicting the pre & post moonsoon data from 2022)

```python
target = df[['Pre-monsoon_2022 (meters below ground level)', 'Post-monsoon_2022 (meters below ground level)']]
```

Splitting the data into training and testing datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
#Currently using 70-30 split
model = LinearRegression()

model.fit(X_train, y_train)
```

Evaluation of the model

```python

score = model.score(X_test, y_test)

print(f'The model score is {score}')
```

```python
predictions = model.predict(X_test)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values for Pre-monsoon and Post-monsoon 2022')
limits = [np.min([plt.xlim(), plt.ylim()]),
          np.max([plt.xlim(), plt.ylim()])]
plt.plot(limits, limits, color='red')  # plot the y = x line

plt.savefig('actual_vs_predicted.png')
plt.show()
```
